# Remaining tasks SNO-STO manuscript

## 1) Check dynamical stability (phonons) of STO

**Context:** 
For a non-layered perovskite one would (for the same A site) expect more rotations for a bigger B site according to the Goldschmidt tolerance factor (https://en.wikipedia.org/wiki/Goldschmidt_tolerance_factor). Somewhat counterintuitively Nb and Ta are the same size (http://abulafia.mt.ic.ac.uk/shannon/ptable.php) even though one would expect Ta to be larger (5d) than Nb (4d). Nevertheless it is puzzling why SNO shows strong rotations (Fig. 1), while STO does not, questioning the structures computed thus far.

**Calculations/approach:**
Using phonon calculations (phonopy) check if STO contains rotational (or other) instabilites and make certain the investigated structures shown in Fig. 1 are sensible to use in a 0K DFT calculation. If we find rotations, we will have to rebuild surfaces and rerun the whole analysis.





## 2) Alternative mechanism on the (100) surface

On (100) check the steps of an alternative O-O dimer desorption and healing mechanisms and if not spontaneous NEB barrier for O migration into the dimer vacancy.

**Context:** 
On the (100) surface Maria considered the dissociated OOH adsorbate as shown in Fig. 4b. She then oxidized the OH fragment and considered O2 to be formed from recombination of two of these O adsorbates. It is however possible that the O2 dimer fragment could desorb, either before or after oxidation of the OH fragment, the resulting vacancy then being filled by the O (oxidized OH).

**Calculations/approach:** 
Compute intermediate structures of the dissociated state with an OH and a dimer vacancy and with an O and a dimer vacancy. Should healing of the vacancy by the OH/O not be spontaneous, perform a NEB to get the barrier for the healing process. Recompute the free energy diagram for these alternative two mechanisms (oxidation first then desorption or desorption and the oxidation) and check if the overpotentials are lower.

## 5) Cut (010) surfaces such that they expose undercoordinated transition-metal sites

**Context:** 
In figure 2 the (010) surfaces do not expose undercoordinated transition-metal ions, limiting their potential reactivity. On the other hand strong TM-O bonds need to be broken to expose TM sites, potentially leading to high surface energies.

**Calculations/approach:** 
Cut the (010) surfaces deeper such that Ta and Nb are undercoordinated and compare surface energies with the structures shown in figure 2. If the surface energies are lower with exposed Nb and Ta, rerun Norskov analysis.

## 7) Formation of stepped/grooved surfaces within Rong and Kolpak model

**Context:** In the 2020 Chem Mat paper (https://doi.org/10.1021/acs.chemmater.9b02294) we calculated the likelihood of formation of grooved/stepped surfaces according to a thermodynamic approach by Rong and Kolpak. This yields the stability of such non stoichiometric surfaces compared to stoichiometric perfect ones as a function of pH and potential, considering the dissolution and formation of the most stable dissolved species under these conditions. This has not been done here and the steps/grooves appear somewhat unjustified.

**Calculations/approach:** 
Analyze the relative stability of the stepped/grooved surfaces using the thermodynamic model. I believe Maria did this in Excel but you may want to choose a fancier approach to solve the equations more easily.
